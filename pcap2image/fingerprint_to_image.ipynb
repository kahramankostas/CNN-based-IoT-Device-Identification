```python
from sklearn import preprocessing
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
def create_df(x):
    df=pd.read_csv("hybrid.csv") 
    df_1= df[df['Label']==x]
    df_1 = df_1.sample(n=25)
    #print(df_1)
    #result=result.reindex(np.random.permutation(result.index))
    return df_1
```

```python

devices=[' Aria',
 ' D-LinkCam',
 ' D-LinkDayCam',
 ' D-LinkDoorSensor',
 ' D-LinkHomeHub',
 ' D-LinkSensor',
 ' D-LinkSiren',
 ' D-LinkSwitch',
 ' D-LinkWaterSensor',
 ' EdimaxCam',
 ' EdimaxPlug1101W',
 ' EdimaxPlug2101W',
 ' EdnetCam',
 ' EdnetGateway',
 ' HomeMaticPlug',
 ' HueBridge',
 ' HueSwitch',
 ' iKettle2',
 ' Lightify',
 ' MAXGateway',
 ' SmarterCoffee',
 ' TP-LinkPlugHS100',
 ' TP-LinkPlugHS110',
 ' WeMoInsightSwitch',
 ' WeMoLink',
 ' WeMoSwitch',
 ' Withings']
flag=1
for i in devices:
    #df=create_df(i)
    if flag:
        df = create_df(i)
        flag=0
    else:
        df = pd.concat([df,create_df(i)])
#df.to_csv('25sample.csv', index=False)
```
